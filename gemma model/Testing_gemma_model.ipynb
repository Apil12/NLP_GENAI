```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Jevvan123/Gemma_finetuned_model")
model = AutoModelForCausalLM.from_pretrained("Jevvan123/Gemma_finetuned_model")
```

```python
# User input question
user_question = "Are your products ethically sourced and manufactured?"

que =[{"role": "user", "content": user_question}]

formatted_question=tokenizer.apply_chat_template(que, tokenize=False)

# Send formatted question to the model for prediction
device = "cuda:0"

inputs = tokenizer.encode(formatted_question , add_special_tokens=False, return_tensors="pt")

outputs = model.generate(input_ids=inputs.to(model.device), max_new_tokens=150)
conversation=tokenizer.decode(outputs[0],skip_special_tokens=True)

que=conversation.split("model")[0][4:].replace("\n","")
ans=conversation.split("model")[1].replace("\n","")

print(f"Question:{que}\n\nAnswer:{ans}")
```
